```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

# Read data from csv file 

```python
stop_file = "static/data/clean_stop_data.csv"
stop_df=pd.read_csv(stop_file)
stop_df.head(1)
```

## Remove all citations with value 'NO'

```python
line = stop_df[stop_df.citationIssued != 'NO']
line.head(2)
```

## Add citation count column for graphing

```python
line['citationCnt'] = line.groupby('neighborhood')['citationIssued'].transform('count')
line
```

## Keep only relevant columns

```python
lines_df = line.drop(['OBJECTID', 'responseDate', 'responseDow','lat', 'long', 'gender'], axis=1)
lines_df.head()
```

## Write out clean csv file

```python
## Clean csv file to be saved in GitHub repository
lines_df.to_csv("static/data/clean_citation_data.csv", index=False,header=True)
```
