# Retrieve data from api and store in csv file


```python
import requests
import json
import pandas as pd
import datetime
```

```python
my_url="https://services.arcgis.com/afSMGVsC7QlRK1kZ/arcgis/rest/services/Police_Stop_Data/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json"
response = requests.get(my_url).json()
```

```python
response['features']
```

```python
stop_data=[record['attributes'] for record in response['features']]
```

format responseDate from number to date

```python
for item in stop_data:
    eftd_date=float(item['responseDate'])/1000.0
    item['responseDate']=datetime.datetime.fromtimestamp(eftd_date).strftime('%Y-%m-%d %H:%M:%S.%f')
        
```

```python
stop_df=pd.DataFrame(stop_data)
stop_df.tail()
```

```python
stop_df.to_csv("copstop/static/data/Police_Stop_data2.csv", index=False,header=True)
```
