```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
stop_file = "static/data/Police_Stop_data.csv"
stop_df=pd.read_csv(stop_file)
stop_df.head()
```

```python
precinct=stop_df.groupby(["policePrecinct"]).count()
#neighbor_sort=neighbor.sort_values("policePrecinct")
precinct
```

```python
neighbor=stop_df.groupby(["policePrecinct","neighborhood"]).count()
#neighbor_sort=neighbor.sort_values("policePrecinct")
neighbor
```

```python
#what gender values are present ?
gender=stop_df.groupby(["gender"]).count()
gender
```

```python
      #select needed columns
select_df=stop_df.loc[:,["OBJECTID", "responseDate","gender","neighborhood" ]]

# set all non-binary gender values to Other
     # fill in null values with Other
dow_df=select_df.fillna("Other")
      #fill in Unknown vaue with Other
dow_df.loc[dow_df.gender=="Unknown","gender"] = "Other"
      # fill in Gender Non-Conforming values with other
dow_df.loc[dow_df.gender =="Gender Non-Conforming","gender"] = "Other"
dow_df.head()
```

```python
#confirm only 3 gender values
gender=dow_df.groupby(["gender"]).count()
gender
```

```python
#parse date
date_list= [datetime.strptime( entry,"%Y-%m-%dT%H:%M:%S.000Z") for entry in dow_df["responseDate"]]
# determine day of week
dow_list= [datetime.strftime(entry,"%w") for entry in date_list]
#add dow column to dow_df
dow_df.loc[:,"dow"]=dow_list
dow_df.head()
```

```python
dow_df.to_csv("static/data/dow_bar.csv", index=False,header=True)
```

```python
# get count of genders by neighborhood and dow
group_df=dow_df.groupby(["neighborhood","dow","gender"]).count()
group_df.head()
```

```python
# remove indexing rename columns and reduce columns
new_group=group_df.reset_index()
next_group=new_group.rename(columns={"OBJECTID":"genderCount"})
last_group=next_group[["neighborhood","dow","gender","genderCount"]]
last_group.head()
```
