```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

# read in file with population data

```python
data_file = "static/data/MSP_CSV_2012_2016.csv"
data_df=pd.read_csv(data_file)
data_df.head(10)
```

# Select desired data, reduce to needed columns and rename for convenience

```python
#neighborhood data
data_df=data_df[data_df["layer"]=='hood']

#minneapolis neighborhoods
data_df=data_df.loc[data_df["City"] =="Minneapolis"]    

#reduce columns and rename
pop_df=data_df[["geography","Total population - Total Population - count","Total population - Total Population - margin"]]  
pop_df.columns=["neighborhood","totalPop","margin"]
   
```

# Set neighborhood as index and fix name differences between data files

```python
# sort and set index to neighborhood
pop_df=pop_df.sort_values("neighborhood")
pop_df=pop_df.set_index("neighborhood")
```

```python
# convert to numeric
pop_df[["totalPop", "margin"]] = pop_df[["totalPop", "margin"]].apply(pd.to_numeric)
# convert to int
pop_df=pop_df.astype('int32')
```

```python
pop_df.head()
```

```python
pop_df=pop_df.rename({"Lind- Bohanon":"Lind - Bohanon", "Como (Minneapolis)":"Como"}, axis='index')
pop_df=pop_df.rename({"Cedar- Isles- Dean":"Cedar - Isles - Dean", "Bryn- Mawr":"Bryn - Mawr"}, axis='index')
pop_df=pop_df.rename({"Steven's Square- Loring Heights":"Steven's Square - Loring Heights", "Sumner- Glenwood":"Sumner - Glenwood"}, axis='index')
pop_df=pop_df.rename({"Mid- City Industrial":"Mid - City Industrial", "Nicollet Island- East Bank":"Nicollet Island - East Bank"}, axis='index')
pop_df=pop_df.rename({"Near North":"Near - North", "Prospect Park- East River Road":"Prospect Park - East River Road"}, axis='index')

pop_df.head()
```

```python
pop_df.to_csv("static/data/population_data.csv", index=True,header=True)
```
